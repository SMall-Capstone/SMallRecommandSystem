```python
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
from pandas import Series, DataFrame
import csv

# ┌CSV 로 학습 ==============================================================

## python:default 로 실행

# csv = pd.read_csv('SMall_woman20_Nx5.csv')

# csv_data = csv[["fashion","beauty","stuff","sports","health"]]
# csv_label = csv["Name"]

# train_data, test_data, train_label, test_label = \
# train_test_split(csv_data, csv_label)

# clf = svm.SVC()
# clf.fit(train_data, train_label)
# pre = clf.predict(test_data)

# # label 을 분리 / 정의 {행조건 , 열조건}
# result = pd.DataFrame({"label" : test_label, "prediction" : pre})
# print(result[1:26])

# ac_score = metrics.accuracy_score(test_label,pre)
# print("\n정답률 = ", ac_score)

# └============================================================================

# tensorflow
# xy = np.loadtxt('RssiData_(1m,3m,5.6m).csv', delimiter=',', dtype=np.float32)

# # 학습할 데이터 선언
# x_train = xy[:, 0:1] # Rssi 값
# y_train = xy[:, [1]] # 실 거리 


rawData = pd.read_csv('SMall_woman20_Nx5.csv')

x_data = DataFrame(rawData, 
                   columns = ["fashion","beauty","stuff","sports","health"])

y_data = DataFrame(rawData,
                  columns = ['Name'])

# traininglist = [None, 5]
# print(x_data.loc[ :, "fashion" : "health" ]) # DataFrame slice 하는법 


# ┌ 해야할일 : Y 를 더미변수로 바꾼다(원 핫 인코딩)
y_dummy = pd.get_dummies(y_data)
print(y_dummy)

# └ 해야할일 : Y 를 더미변수로 바꾼다(원 핫 인코딩)

# 해야할일 : DataFrame 형을 placeholder 에 넣어야함

# ┌ DataFrame 을 matrix 로 변환 ===========================================

x_mat = x_data.as_matrix()

y_mat = y_dummy.as_matrix()
# y_mat = np.asmatrix(y_mat).T # y 를 2차원 행렬 형태로 변환

print(x_mat)
print(y_mat)

# └ DataFrame 을 matrix 로 변환 ===========================================



# 데이터 분할

# x_train, x_test, y_train, y_test = train_test_split(x_std, y_mat, test_size=0.4, random_state=0)

X = tf.placeholder("float32" , [None, 5])
Y = tf.placeholder("float32", [None, 5])
nb_classes = 5

W = tf.Variable(tf.random_normal([5, nb_classes]), name = 'weight' )
b = tf.Variable(tf.random_normal([nb_classes]), name = 'bias' )

hypo = tf.nn.softmax(tf.matmul(X,W) + b)
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hypo) , axis = 1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    sess.run(optimizer, feed_dict = {X: x_mat, Y: y_mat})
    if step % 200 == 0:
        print(step , sess.run(cost, feed_dict = {X : x_mat, Y : y_mat}))
        


# └=================================================================    

# a = pd.get_dummies(x_data)

# print(a)

# X = tf.placeholder("float32", [None, 5])
# Y = tf.placeholder("float32", [None, 1])
# nb_classes = 1

# W = tf.Variable(tf.random_normal([5, nb_classes]), name = 'weight' )
# b = tf.Variable(tf.random_normal([nb_classes]), name = 'bias' )

# hypo = tf.nn.softmax(tf.matmul(X,W) + b)
# cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hypo) , axis = 1))
# optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# sess = tf.Session()
# sess.run(tf.global_variables_initializer())

# for step in range(2001):
#     sess.run(optimizer, feed_dict = {X: x_data, Y: y_data})
#     if step % 200 == 0:
#         print(step , sess.run(cost, feed_dict = {X : x_data, Y : y_data}))
        
# a = sess.run(hypo, feed_dict = {X : [[1, 8, 7, 9, 0]]})
# print(a, sess.run(tf.argmax(a,1)))
# ==========================================================
```
